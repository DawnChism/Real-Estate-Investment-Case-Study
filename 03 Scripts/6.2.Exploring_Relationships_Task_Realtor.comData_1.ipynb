#### 3.There are several variables that have a strong correlation. 
#### With a 0.90 coefficient for average listing price to median listing price indicates a very strong positive correlation between these variables and that the average list price and the median list price are very close together. 
#### With a 0.90 coefficient for total listings to new listings indicates a very strong positive correlation between these variables and shows that a large percentage of the total listings are new listings. 
#### With a 0.90 coefficient for total listings to price reduced indicates a very strong positive correlation between these variables and shows that a large percentage of total listings incur a price reduction. 

#### With a 0.80 coefficient for new listings to active listings indicates a very strong positive correlation between these variables and shows that a large percentage of active listings are new listings.
#### With a 0.80 coefficient for price reduced to active listings indicates a very strong positive correlation between these variables and shows that a large percentage of active listings have experienced a price reduction. 
#### With a 0.80 coefficient for price reduced to new listings indicates a very strong positive correlation between these variables and shows that a large percentage of active listings have experienced a price reduction. 
#### With a 0.80 coefficient for pending listing to new listings indicates a very strong positive correlation between these variables and shows that a large percentage of new listings become pending listings. 
#### With a 0.80 coefficient for total listings to pending listings indicates a very strong positive correlation between these variables and shows that a large percentage of total listings become pending listings. 

#### With 0.60 coefficient for pending listings to active listings indicates a medium-to-strong positive correlation shows that a more than half of all active listings go pending. 
#### With 0.60 coefficient for pending listings to price increases indicates a medium-to-strong positive correlation shows that a large amount of listings that go pending experience a price increase. This could be from homes selling over list price, price increase from repairs or maintenace to the house.
#### With 0.60 coefficient for pending listings to price reduced indicates a medium-to-strong positive correlation shows that a large amount of pending listings experience a price reduction.

#### With 0.5 coefficient for price increase to new listings indicates a moderate positive correlation shows that a moderate amount of new listings experience a price increase for one reason or another.
#### With 0.5 coefficient for price reduced to price increase indicates a moderate positive correlation shows that a moderate amount of homes that experience a price decrease also experience a price increase for one reason or another. Again, it could be due to the house selling for over list price.

#### With 0.4 coefficient for price increase to active listings indicates a a moderate positive correlation shows that a moderate amount of active listings experience a price increase. 

#### With -0.1 coefficient for total listings to median days on market indicates a weak negative correlation and could be interprested as the more total listings there are the higher the median days on market. 

#### With -0.2 coefficient for pending listings to median days on market indicates a weak negative correlation and could be interprested as less  pending listings equates to higher median days on market. 

#### With -0.3 coefficient for new listings to median days on market indicates a weak negative correlation and could be interprested as less new listings equates to higher median days on market.
